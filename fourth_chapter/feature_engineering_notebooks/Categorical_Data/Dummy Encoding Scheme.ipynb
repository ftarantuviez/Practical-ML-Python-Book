# Dummy Encoding Scheme

This what it does is the same than one hot coding scheme, where for each category is created a new feature based in binary values. But here, in diference with OHCS, the features goes from 0 to *m*-1

```python
df = pd.read_csv("../../datasets/pokemon.csv")
```

```python
gen_dummy_features = pd.get_dummies(df["Generation"], drop_first=True)
pd.concat([df[["Name", "Generation"]], gen_dummy_features], axis=1).iloc[164:170]
```

```python
gen_onehot_features = pd.get_dummies(df["Generation"])
gen_dummy_features = gen_onehot_features.iloc[:, :-1]
pd.concat([df[["Name", "Generation"]], gen_dummy_features], axis=1)
```

## Observations

The idea of this scheme is drop one of the categorical values (one of the classes), that that class will be represented as a full 0s columns in the other features.
