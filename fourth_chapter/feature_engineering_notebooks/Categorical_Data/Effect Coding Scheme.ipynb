# Effect Coding Scheme

Is similar to Dummy Coding Scheme, but instead here the category or class dropped is represented by a full columns equal to one

```python
df = pd.read_csv("../../datasets/pokemon.csv")
```

```python
gen_onehot_features = pd.get_dummies(df["Generation"])
gen_effect_features = gen_onehot_features.iloc[:, :-1]
gen_effect_features.loc[np.all(gen_effect_features == 0, axis=1)] = -1
pd.concat([df[["Name", "Generation"]], gen_effect_features], axis=1)
```
