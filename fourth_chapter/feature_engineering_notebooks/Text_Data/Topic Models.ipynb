```python
import pandas as pd
import numpy as np

from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer
```

# Topic Models

This models search to assign an specific general topic

```python
norm_corpus = np.array(['sky blue beautiful', 'love blue beautiful sky',
       'quick brown fox jumps lazy dog', 'brown fox quick blue dog lazy',
       'sky blue sky beautiful today', 'dog lazy brown fox quick'],
      dtype='<U30')
```

```python
tv = TfidfVectorizer(min_df=0., max_df=1., use_idf=False)
matrix_tv = tv.fit_transform(norm_corpus)
matrix_tv = matrix_tv.toarray()
vocab = tv.get_feature_names()
pd.DataFrame(matrix_tv, columns=tv.get_feature_names())
```

```python
lda = LatentDirichletAllocation(n_components=2, max_iter=100, random_state=42)
dt_matrix = lda.fit_transform(matrix_tv)
features = pd.DataFrame(dt_matrix, columns=["Topic 1", "Topic 2"])
features
```

```python
tt_matrix = lda.components_
for topic_weights in tt_matrix:
    topic = [(token, weight) for token, weight in zip(vocab, topic_weights)]
    topic = sorted(topic, key=lambda x: -x[1])
    topic = [item for item in topic if item[1] > .6]
    print(topic)
    print()
```
