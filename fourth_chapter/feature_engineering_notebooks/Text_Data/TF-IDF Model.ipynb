```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
```

# TF-IDF Model

This model search to replace the absolute frequencies and use an specific score to represent the recurrence of each word. This model consider to metrics:

- tf, term frequency
- idf, inverse document frequency

```python
norm_corpus = np.array(['sky blue beautiful', 'love blue beautiful sky',
       'quick brown fox jumps lazy dog', 'brown fox quick blue dog lazy',
       'sky blue sky beautiful today', 'dog lazy brown fox quick'],
      dtype='<U30')
```

```python
tv = TfidfVectorizer(min_df=0., max_df=1., use_idf=True)
tv_matrix = tv.fit_transform(norm_corpus)
tv_matrix = tv_matrix.toarray()

tv_labels = tv.get_feature_names()
pd.DataFrame(tv_matrix, columns=tv_labels)
```
