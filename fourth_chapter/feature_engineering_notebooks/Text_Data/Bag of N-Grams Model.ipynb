```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

# Bag of N-Grams Model

This consists, as equal as Bag of Words Model, in count the frequency of each diferent word in each text of the document. But instead, here the idea is count groups of words, combinations of them. *N* is the size of the groups of combinations.

```python
norm_corpus = np.array(['sky blue beautiful', 'love blue beautiful sky',
       'quick brown fox jumps lazy dog', 'brown fox quick blue dog lazy',
       'sky blue sky beautiful today', 'dog lazy brown fox quick'],
      dtype='<U30')
```

```python
bv = CountVectorizer(ngram_range=(2, 2))
bv_matrix = bv.fit_transform(norm_corpus)
bv_matrix = bv_matrix.toarray()
bv_columns = bv.get_feature_names()
pd.DataFrame(bv_matrix, columns=bv_columns)
```
