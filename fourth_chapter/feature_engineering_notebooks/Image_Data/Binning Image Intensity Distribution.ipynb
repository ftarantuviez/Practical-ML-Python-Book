```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import skimage
from skimage import io
from skimage.color import rgb2gray
```

# Binning Image Intensity Distribution

To see the frequency of intensity in an image, we can bin the intensity values and visualize them in a histogram to whatch where the image have more intesity.

```python
cat = io.imread("../../datasets/cat.png")
dog = io.imread("../../datasets/dog.png")
cgs = rgb2gray(cat)
dgs = rgb2gray(dog)
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cgs)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(dgs)
ax3 = fig.add_subplot(2,2,3)
c_freq, c_bins, c_patches = ax3.hist(cgs.flatten(), bins=30)
ax4 = fig.add_subplot(2,2,4)
d_freq, d_bins, d_patches = ax4.hist(dgs.flatten(), bins=30)
```
