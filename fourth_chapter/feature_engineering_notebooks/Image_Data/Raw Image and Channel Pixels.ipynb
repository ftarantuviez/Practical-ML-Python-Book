# Raw Image and Channel Pixels

Each image can be represented as a two dimensional array, where the dimensions indicates the dimensions of the image in pixels. But color images have three components called channels which depicts the colors of it. So in this cases, the image is represented by a three dimensional array (m, n, c), where m indicates the number of rows, n the number of columns and c the channels used.
We can handle these matrix as numpy arrays, and change the colors, the size, etc.

```python
cat = io.imread("../../datasets/cat.png")
dog = io.imread("../../datasets/dog.png")
df = pd.DataFrame(["Cat", "Dog"], columns=["Image"])

print(cat.shape, dog.shape)
```

Here we can see that is an image of 168x300 pixels, and have the three channels of colors.

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(cat)
ax1.set_title("Cat")
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(dog)
ax2.set_title("Dog")
```

```python
dog_r = dog.copy()
dog_r[:,:,1] = dog_r[:,:,2] = 0 # set G,B pixels = 0
dog_g = dog.copy()
dog_g[:,:,0] = dog_r[:,:,2] = 0  # set R,B pixels = 0
dog_b = dog.copy()
dog_b[:,:,0] = dog_b[:,:,1] = 0 # set R,G pixels = 0

plot_image = np.concatenate((dog_r, dog_g, dog_b), axis=1)
plt.figure(figsize=(10,4))
plt.imshow(plot_image)
```
