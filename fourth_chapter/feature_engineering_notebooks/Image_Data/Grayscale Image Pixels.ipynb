```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import skimage
from skimage import io
from skimage.color import rgb2gray
```

# Grayscale Image Pixels

The main idea is to handle images easily reducing the dimensionality of the array.

```python
dog = io.imread("../../datasets/dog.png")
cat = io.imread("../../datasets/cat.png")
```

```python
cat_grayscale = rgb2gray(cat)
dog_grayscale = rgb2gray(dog)

print("Image shape:", cat_grayscale.shape, "\n")

# 2D pixel map
print("2D image pixel map")
print(np.round(cat_grayscale,2), "\n")

# flattened pixel feature vector
print("Flattened pixel map:", (np.round(cat_grayscale.flatten(), 2)))
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(cat_grayscale)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(dog_grayscale)
```
